Sum of digits:Given an integer N,print the sum of its digits.

```python
n = int(input("Enter a number: "))
total = 0
while n > 0:
    total += n % 10
    n //= 10
print("Sum of digits:", total)
```

Reverse a number: Reverse the given integer without converting to string

```python
n = int(input("Enter a number: "))
rev = 0
while n > 0:
    digit = n % 10
    rev = (rev * 10) + digit
    n //= 10
print("Reversed number:", rev)
```

Armstrong Number: Check whether a number is Armstrong or not

```python
n = int(input("Enter a number: "))
temp = n
digits = len(str(n))
sum_pow = 0

while temp > 0:
    digit = temp % 10
    sum_pow += digit ** digits
    temp //= 10

if n == sum_pow:
    print(f"{n} is an Armstrong number")
else:
    print(f"{n} is not an Armstrong number")
```

Fibonacci Series: Print first N Fibonacci numbers.

```python
n = int(input("Enter count: "))
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

Prime Count: Count how many prime numbers exist from 1 to N.

```python
n = int(input("Enter N: "))
prime_count = 0

for num in range(2, n + 1):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        prime_count += 1

print(f"There are {prime_count} primes between 1 and {n}")
```

Factorial Function: Write a function to return factorial of a number.

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Usage
num = int(input("Enter a number: "))
print(f"Factorial of {num} is {factorial(num)}")
```

Largest of Three: Create a function to find the largest of three numbers.

```python
def find_largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# Usage
print("Largest is:", find_largest(10, 25, 15))
```

Count Vowels: Create a function to count vowels in a string.

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

# Usage
txt = input("Enter a string: ")
print(f"Number of vowels: {count_vowels(txt)}")
```

Simple Calculator: Create menu-driven calculator using functions.

```python
def add(x, y): return x + y
def subtract(x, y): return x - y
def multiply(x, y): return x * y
def divide(x, y): return x / y if y != 0 else "Error! Division by zero."

def calculator():
    print("\n--- Simple Calculator ---")
    print("1. Add\n2. Subtract\n3. Multiply\n4. Divide\n5. Exit")

    while True:
        choice = input("\nEnter choice (1-5): ")
        if choice == '5':
            print("Exiting...")
            break

        if choice in ('1', '2', '3', '4'):
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))

            if choice == '1': print("Result:", add(num1, num2))
            elif choice == '2': print("Result:", subtract(num1, num2))
            elif choice == '3': print("Result:", multiply(num1, num2))
            elif choice == '4': print("Result:", divide(num1, num2))
        else:
            print("Invalid Input")

# Run the calculator
calculator()
```

* Random Guess Game: Generate random number (1-100) and build guessing game using loop.

```python
import random

# Generate a random number between 1 and 100
number = random.randint(1, 100)

print("🎯 Welcome to the Random Guess Game!")
print("Guess a number between 1 and 100")

while True:
    guess = int(input("Enter your guess: "))

    if guess < number:
        print("Too low! Try again.")
    elif guess > number:
        print("Too high! Try again.")
    else:
        print("🎉 Congratulations! You guessed the correct number!")
        break
```

Student Management System: Create a menu-driven program that allows users to add
student details, view all students, search by roll number, and calculate average marks. Use loops,
functions, and modular programming concepts.

```python
# Student Management System

students = []

# Function to add student
def add_student():
    roll = input("Enter Roll Number: ")
    name = input("Enter Name: ")
    marks = float(input("Enter Marks: "))

    student = {
        "roll": roll,
        "name": name,
        "marks": marks
    }

    students.append(student)
    print("Student added successfully!\n")


# Function to view all students
def view_students():
    if len(students) == 0:
        print("No students found.\n")
        return

    print("\nStudent List:")
    for student in students:
        print("Roll:", student["roll"],
              "| Name:", student["name"],
              "| Marks:", student["marks"])
    print()


# Function to search student by roll number
def search_student():
    roll = input("Enter Roll Number to search: ")

    for student in students:
        if student["roll"] == roll:
            print("Student Found!")
            print("Name:", student["name"])
            print("Marks:", student["marks"], "\n")
            return

    print("Student not found.\n")


# Function to calculate average marks
def calculate_average():
    if len(students) == 0:
        print("No students available to calculate average.\n")
        return

    total = 0
    for student in students:
        total += student["marks"]

    average = total / len(students)
    print("Average Marks:", average, "\n")


# Main Menu Loop
while True:
    print("===== Student Management System =====")
    print("1. Add Student")
    print("2. View All Students")
    print("3. Search Student by Roll Number")
    print("4. Calculate Average Marks")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        calculate_average()
    elif choice == "5":
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice! Please try again.\n")
```
